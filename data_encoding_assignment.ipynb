### Q1. What is data encoding? How is it useful in data science?

Data encoding is the process of converting categorical data into numerical format so that machine learning algorithms can understand it. 
It is useful because most algorithms work with numbers, and encoding helps in improving model performance and interpretability.

### Q2. What is nominal encoding? Provide an example of how you would use it in a real-world scenario.

Nominal encoding is used for categorical variables that have no natural order. 
Example: Encoding colors like Red, Blue, Green into numbers or one-hot vectors in a product recommendation system.

### Q3. In what situations is nominal encoding preferred over one-hot encoding? Provide a practical example.

Nominal encoding (label encoding) is preferred when:
- Number of categories is very large
- Model can handle numeric relationships (like tree-based models)

Example: Encoding city names in a large dataset with thousands of cities.

### Q4. Dataset with 5 unique categorical values - which encoding?

I would use One-Hot Encoding because:
- Only 5 categories (small size)
- Avoids false ordinal relationships
- Works well with most machine learning algorithms

### Q5. Dataset with 1000 rows and 5 columns (2 categorical)

New columns = 4 + 3 = 7 columns

Total columns after encoding:
= 3 numerical + 7 encoded = 10 columns

### Q6. Animal dataset encoding choice

I would use One-Hot Encoding because:
- Features like species, habitat, diet are nominal
- No inherent order exists
- Ensures model does not assume relationships between categories

### Q7. Telecom churn project encoding

Steps:
1. Identify categorical features (gender, contract type)
2. Apply One-Hot Encoding for nominal variables
3. Use Label Encoding if binary (e.g., gender)
4. Normalize numerical features (age, charges, tenure)
5. Combine all features into final dataset
6. Train model on encoded data
